```python
#izracunaj spodnjo in zgornjo mejo koze. Parametri so slika in koordinati skatle
def doloci_barvo_koze(slika, levo_zgoraj, desno_spodaj):
    print("Poteka določanje barve slike")
    #print("Info: Koordinata škatle levo zgoraj:",levo_zgoraj)
    #print("Info: Koordinata škatle desno spodaj:",desno_spodaj)
    #spremenljivke, na katere bomo preverjali posamezne piksle
    spodnja_meja_B = 255
    zgornja_meja_B = 0
    spodnja_meja_G = 255
    zgornja_meja_G = 0
    spodnja_meja_R = 255
    zgornja_meja_R = 0
    x = levo_zgoraj[0]
    y = levo_zgoraj[1]
    #horizontalni pravokotnik
    #Max_x = desno_spodaj[0]
    #Max_y = desno_spodaj[1]
    #vertiklani pravokotnik
    Max_y = desno_spodaj[0]
    Max_x = desno_spodaj[1]
    print("Info: Koordinata škatle levo zgoraj: [{},{}]".format(x,y))
    print("Info: Koordinata škatle desno spodaj: [{},{}]".format(Max_x,Max_y))
    
    #skupne vrednosti posameznega piksla (za povprečje)
    array_px_B = []
    array_px_G = []
    array_px_R = []

    #pojdi po x osi
    for i in range(x,Max_x):
        #pojdi po y osi
        for j in range(y,Max_y):
            px=slika[i,j] #piksel na trenutni lokaciji (BGR)
            #print("Vrednost piksla barvne slike na lokaciji [{},{}] = {}".format(i,j,px))
            
            '''
            avg_px=int(px[0])+int(px[1])+int(px[2])/3 #izračunaj povprečje trenutnega piksla
            print("Povprečje piksla barvne slike na lokaciji [{},{}] = {}".format(i,j,avg_px))
            
            print("Avg {} Zgornja meja {}".format(avg_px,zgornja_meja))
            avg_px = np.array(avg_px)
            zgornja_meja = np.array(zgornja_meja)
            spodnja_meja = np.array(spodnja_meja)
            if (avg_px>zgornja_meja).all():
                zgornja_meja=copy.deepcopy(avg_px)
            if (avg_px<spodnja_meja).all():
                spodnja_meja=copy.deepcopy(avg_px)
            '''
            
            array_px_B.append(px[0])
            array_px_G.append(px[1])
            array_px_R.append(px[2])

            #standaradni odklon - polje, samo dodajaš piksle

            if (px[0]>zgornja_meja_B):
                zgornja_meja_B=px[0]
            if (px[0]<spodnja_meja_B):
                spodnja_meja_B=px[0]
            if (px[1]>zgornja_meja_G):
                zgornja_meja_G=px[1]
            if (px[1]<spodnja_meja_G):
                spodnja_meja_G=px[1]
            if (px[2]>zgornja_meja_R):
                zgornja_meja_R=px[2]
            if (px[2]<spodnja_meja_R):
                spodnja_meja_R=px[2]
            
            #print("Vrednost piksla na lokaciji [{},{}] = {}".format(i,j,px))
            #print("Vrednost piksla na lokaciji [{},{}]={} , povprečje pa {}".format(i,j,px,avg_px))
    
    avg_B = sum(array_px_B)/len(array_px_B)
    avg_G = sum(array_px_G)/len(array_px_G)
    avg_R = sum(array_px_R)/len(array_px_R)
    
    '''
    print("Vsi B piksli {}".format(array_px_B))
    print("Vsi G piksli {}".format(array_px_G))
    print("Vsi R piksli {}".format(array_px_R))
    
    print("Povprečje B pikslov {}".format(avg_B))
    print("Povprečje G pikslov {}".format(avg_G))
    print("Povprečje R pikslov {}".format(avg_R))

    print("Standardni odklon B pikslov {}".format(np.std(array_px_B)))
    print("Standardni odklon G pikslov {}".format(np.std(array_px_G)))
    print("Standardni odklon R pikslov {}".format(np.std(array_px_R)))
    '''
    
    #print("Spodnja in zgornja meja kože sta {} {}".format(spodnja_meja,zgornja_meja))        
    #zgornja_meja_koze = zgornja_meja
    #spodnja_meja_koze = spodnja_meja
    '''
    zgornja_meja_B = int((sum_B)/(i*j)) + np.std([zgornja_meja_B,spodnja_meja_B])
    zgornja_meja_G = int((sum_G)/(i*j)) + np.std([zgornja_meja_G,spodnja_meja_G])
    zgornja_meja_R = int((sum_R)/(i*j)) + np.std([zgornja_meja_R,spodnja_meja_R])
    spodnja_meja_B = int((sum_B)/(i*j)) - np.std([zgornja_meja_B,spodnja_meja_B])
    spodnja_meja_G = int((sum_G)/(i*j)) - np.std([zgornja_meja_G,spodnja_meja_G])
    spodnja_meja_R = int((sum_R)/(i*j)) - np.std([zgornja_meja_R,spodnja_meja_R])
    '''
    
    zgornja_meja_B = avg_B + np.std(array_px_B)
    zgornja_meja_G = avg_G + np.std(array_px_G)
    zgornja_meja_R = avg_R + np.std(array_px_R)
    spodnja_meja_B = avg_B - np.std(array_px_B)
    spodnja_meja_G = avg_G - np.std(array_px_G)
    spodnja_meja_R = avg_R - np.std(array_px_R)      
    
    zgornja_meja_koze = (int(zgornja_meja_B), int(zgornja_meja_G), int(zgornja_meja_R))
    spodnja_meja_koze = (int(spodnja_meja_B), int(spodnja_meja_G), int(spodnja_meja_R))
    return (spodnja_meja_koze, zgornja_meja_koze)

#pridobi sliko in jo zmanjsaj na 220x340
def zmanjsaj_sliko(slika):
    #zmanjsaj pridobljeno sliko na 340x220 (obdrzi aspect ratio) pikslov z hitro inpterploacijsko metodo
    slika_smaller = cv2.resize(slika, (340,220), interpolation=cv2.INTER_AREA)
    return slika_smaller

#vrni koordinate skatle, ki se najboljse ujema. Parametri okna: float odstotek celotne slike
def obdelaj_sliko(slika, okno_sirina, okno_visina, barva_koze_spodaj, barva_koze_zgoraj):
    x = 0
    y = 0
    #horizontalni pravokotnik
    #Max_x = slika.shape[0]
    #Max_y = slika.shape[1]
    #vertiklani pravokotnik
    Max_x = slika.shape[1]
    Max_y = slika.shape[0]
    match = 0 #število elementov, ki se ujemajo
    print("Info: Dimenzije slike: [{},{}]".format(Max_x,Max_y))

    #inicializiraj koordinate skatle
    levo_zgoraj_x=x
    levo_zgoraj_y=y
    desno_spodaj_x=int(Max_x*okno_visina)
    desno_spodaj_y=int(Max_y*okno_sirina)
    print("Info: Koordinate škatle levo zgoraj: [{},{}]".format(levo_zgoraj_x,levo_zgoraj_y))
    print("Info: Koordinate škatle desno spodaj: [{},{}]".format(desno_spodaj_x,desno_spodaj_y))
    levo_zgoraj_x_idealno=0
    levo_zgoraj_y_idealno=0
    desno_spodaj_x_idealno=0
    desno_spodaj_y_idealno=0


    #škatlo premikaj čez celotno sliko, piksel po piksel
    while (desno_spodaj_x<Max_x) & (desno_spodaj_y<Max_y):
        podslika = slika[levo_zgoraj_x:desno_spodaj_x, levo_zgoraj_y:desno_spodaj_y]
        #print("Info: Koordinate podslike: x=[{},{}] y=[{},{}]".format(levo_zgoraj_x,desno_spodaj_x, levo_zgoraj_y,desno_spodaj_y))
        match_new = prestej_piksle_z_barvo_koze(podslika,barva_koze_spodaj,barva_koze_zgoraj)
        print("Info: Old match: {}  New match: {}".format(match,match_new))
        
        #if(match_new>match):
        if(match_new>=match):
            match=copy.deepcopy(match_new)
            print("Poiskali boljše ujemanje na koordinatah: x=[{},{}] y=[{},{}]".format(levo_zgoraj_x,desno_spodaj_x, levo_zgoraj_y,desno_spodaj_y))
            levo_zgoraj_x_idealno = copy.deepcopy(levo_zgoraj_x)
            levo_zgoraj_y_idealno = copy.deepcopy(levo_zgoraj_y)
            desno_spodaj_x_idealno = copy.deepcopy(desno_spodaj_x)
            desno_spodaj_y_idealno = copy.deepcopy(desno_spodaj_y)
        levo_zgoraj_x+=1
        desno_spodaj_x+=1 
        levo_zgoraj_y+=1
        desno_spodaj_y+=1

    #slika[levo_zgoraj_x_idealno,levo_zgoraj_y_idealno]=[255,0,255]
    #slika[desno_spodaj_x_idealno,desno_spodaj_y_idealno]=[255,0,255]
    #plt.imshow(cv2.cvtColor(slika,cv2.COLOR_BGR2RGB)) #pikaži v RGB formatu (prvotne slike v BGR)
    levo_zgoraj_idealno = (levo_zgoraj_x_idealno, levo_zgoraj_y_idealno)
    desno_spodaj_idealno = (desno_spodaj_x_idealno, desno_spodaj_y_idealno)

    print("Najboljše ujemanje na koordinatah: x=[{},{}] y=[{},{}]".format(levo_zgoraj_x_idealno,desno_spodaj_x_idealno, levo_zgoraj_y_idealno,desno_spodaj_y_idealno))
    return (levo_zgoraj_idealno, desno_spodaj_idealno)

#prestej st. slikovnih elementov v podsliki, ki pripadajo barvi koze (so v meji)
def prestej_piksle_z_barvo_koze(podslika, barva_koze_spodaj, barva_koze_zgoraj):
    '''
    while(True): #prikazuj dokler uporabnik ne pritisne gumb
            cv2.imshow('Podslika',podslika) #prikazi okvir v novem oknu
            if cv2.waitKey(1) & 0xFF == ord('q'): #pritisni q in zapri okno
                cv2.destroyAllWindows() #brez tega, program crash-a
                break
    '''
    #print("Info: Koordinate podslike: [{},{}]".format(podslika.shape[0],podslika.shape[1]))
    #plt.imshow(cv2.cvtColor(podslika,cv2.COLOR_BGR2RGB)) #pikaži v RGB formatu (prvotne slike v BGR)

    #pojdi po sliki in računaj spodnjo/zgornjo mejo
    x=0
    y=0
    Max_x=podslika.shape[0]
    Max_y=podslika.shape[1]
    match = 0 #število elementov, ki se ujemajo

    #pojdi po y osi
    for i in range(x,Max_x):
        #pojdi po y osi
        for j in range(y,Max_y):
            px=podslika[i,j] #piksel na trenutni lokaciji (BGR)
            #print("Vrednost piksla barvne slike na lokaciji [{},{}] = {}".format(i,j,px))

            #poglej, če je trenutni piksel v mejah kože
            if (px>barva_koze_spodaj).all() & (px<barva_koze_zgoraj).all():
                match=match+1 #število pikslov, ki se ujemajo se poveča
    return match
```
